#### Open the intake Catalog

```python
import intake
catalog  = intake.from_yaml_file('https://mastapp.site/intake/catalog.yml')
list(catalog.entries)
```

#### List All Shots

```python
shots_df = catalog.index_shots.read()
shots_df
```

#### List All Sources

```python
sources_df = catalog.index_sources.read()
sources_df
```

#### List all signals Given a Shot Number

```python
signals_df = catalog(shot_id=30420).index_signals.read()
signals_df
```

#### Open and plot shot data. 

Here we are visualising some plasma current data.

```python
import matplotlib.pyplot as plt
dataset = catalog(urlpath="s3://mast/level1/shots/30420.zarr/amc").level1_sources.read()
plt.plot(dataset['time'], dataset['plasma_current'])
```
