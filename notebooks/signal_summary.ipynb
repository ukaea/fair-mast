Overview of signal status

Get a list of the units

```python
ureg = UnitRegistry()
ureg.define('@alias tesla = Tesla')
ureg.define('@alias watt = Watts')

unit_map = json.load(Path('~/mast-data/mappings/units.json').expanduser().open('r'))

def parse_unit(name, mode='compact'):
    name = unit_map[name] if name in unit_map else name
    try:
        quantity = ureg.parse_expression(name)
        if mode == 'compact':
            return "{:~C}".format(quantity.u)
        else:
            return "{}".format(quantity.u)
    except:
        return ''

units_summary = df[['name', 'label', 'description', 'units']].drop_duplicates().reset_index(drop=True)
units_summary['description'] = units_summary.description.astype(str)
units_summary = units_summary.rename(dict(units='old_units'), axis=1)
units_summary['new_units'] = units_summary['old_units'].map(parse_unit)
units_summary['long_name'] = units_summary['old_units'].map(partial(parse_unit, mode='verbose'))
units_summary.to_parquet('~/mast-data/mappings/units.parquet')
units_summary
show(units_summary, style="table-layout:auto;width:50%;float:left")
```

Look at the unique dimension names, see what we can combine:

```python
dims = df[['name', 'dimensions']]
dims.loc[:, 'dimensions'] = dims.dimensions.map(lambda x: tuple(x))
dims = dims.drop_duplicates()
dim_names = []
for key, item in dims.iterrows():
    for dim in item.dimensions:
        dim_names.append((item['name'], dim, dim))

dim_map = json.load(Path('~/mast-data/mappings/dimensions.json').expanduser().open('r'))

def parse_dimensions(dim_name):
    return dim_map.get(dim_name, dim_name)

dims = pd.DataFrame(dim_names, columns=['name', 'old_dimension', 'new_dimensions'])
dims['new_dimensions'] = dims['new_dimensions'].map(parse_dimensions)
dims.to_parquet('~/mast-data/mappings/dimensions.parquet')
dims
```
